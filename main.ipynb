```python
#import libraries
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50, 60, 70])
print(arr)
print("\n Average number:", arr.mean())
```

```python
#indexing
print(arr[3])
print(arr[-2])
print(arr[6])
print(arr[-4])
```

```python
#slicing is a method of assessing value between the index range of that value
print(arr[1:6])
print(arr[6:])
print(arr[:7])
print(arr[-6:-1])
print(arr[7:])
print(arr[-1:-6])
```
